```python
def get_match_data(region, match_id, api_key ):
    api_url = (
        "https://" +
        region + 
        ".api.riotgames.com/tft/match/v1/matches/" + 
        match_id + 
        "?api_key=" +
        api_key
    ) 
    
    resp = requests.get(api_url)
    data = resp.json()
    return data
```

```python
def tft_profile(puuid, match_data):
    try:
        part_index = match_data['metadata']['participants'].index(puuid)
        return match_data['info']['participants'][part_index]['placement']
    except Exception as e:
        print(match_data)
    
```

```python
def get_matches(region, puuid, count, api_key):
    api_url = (
        "https://" +
        region +
        ".api.riotgames.com/tft/match/v1/matches/by-puuid/" + 
        puuid +
        "/ids?" +
        "start=0&" + 
        "count=" +
        str(count) +
        "&api_key=" + 
        api_key
    )
        
    resp = requests.get(api_url)
    return resp.json()
  
    
```
